# Отчет по задаче №1 - SQL

## Описание задачи
У компании HeadHunter основные клиенты – это работодатели.<br/>
<br/>
Данные о них хранятся в табличке ***Account***. Там располагается вся самая важная информация: код клиента, название, идентификаторы статусов и т.д.<br/>
<br/>
Работодатели могут приобрести пакет вакансий на нашем сайте (таблица ***Orders***). Затем они могут опубликовать вакансию, использовав один из имеющихся пакетов. Для эффективного использования наших услуг клиентами, мы проводим аналитику и даем рекомендации по работе с нашими сервисами. Для этого нам нужно оценить как клиент пользуется услугами и какую эффективность он получает.<br/>
<br/>
Существует такое понятие, как ***трата вакансии*** (таблица ***Spending***). Это значит, что клиент публикует или обновляет существующую вакансию – в этот момент происходит трата услуги. Одна и та же вакансия может быть один раз опубликована, затем продлена (или поднята в выдаче) неограниченное количество раз – все это будет считаться ***тратой*** вакансии. Пока вакансия висит в выдаче (т.е. она опубликована и еще не ушла в архив – см. дату архивации) соискатели могут на нее ***откликнуться*** (таблица ***VacancyResponse***). Пошаговый пример:<br/>
<br/>
1. Клиент «Ромашка» создает вакансию «Менеджер» и покупает пакет из 195 вакансий стандарт.
1. Далее он публикует Вакансию и из его пакета тратится 1 вакансия/трата. 
1. Через 30 дней он тратит еще одну вакансию из пакета (у него остается 193 в запасе) и вакансия снова поднимается в поисковой выдаче. 
1. До тех пор, пока вакансия не уйдет в архив, дата архивации у нее будет NULL и на нее сможет откликнуться соискатель (т.е. на нее могут поступать отклики)

## Задание: 
<br/>
Необходимо посчитать среднюю стоимость отклика на трату вакансии клиента «Ромашка» за 2021 год:

- Периодом активной траты считать время от начала траты до следующей траты в рамках одной Вакансии. 
- При отсутствии последующей траты – использовать дату архивации вакансии, если вакансия еще не в архиве и следующей траты нет, то берем текущую дату. 
- За среднюю стоимость одного отклика на трату взять сумму всех затрат на вакансию, деленную на сумму всех откликов, полученных на эту вакансию (sum(spent_cost)/sum(response_count))


В результате мы хотим получить скрипт, который выдаст ответ вида: код клиента, название клиента, год траты услуги, средняя стоимость отклика на трату, количество трат, количество откликов, сумма затрат клиента – должна получиться 1 строка!<br/>

Обратите внимание, что в файле приведены ***примеры** заполнения таблиц. Для отладки скриптов ***можно*** придумать собственные данные.<br/>


### Дальше у меня есть описание желаемого конечного результата:

"Необходимо посчитать среднюю стоимость отклика на трату вакансии клиента «Ромашка» за 2021 год"

и

"В результате мы хотим получить скрипт, который выдаст ответ вида: 

- Код клиента
- Название клиента
- Год траты услуги
- Средняя стоимость отклика на трату (sum(spent_cost)/sum(response_count))
- Количество трат
- Количество откликов
- Сумма затрат клиента

должна получиться 1 строка!"

### Получился вот такой скрипт:

``` mysql

with
    (select employer_id from headhunter.Account where Name == 'Ромашка') as employer_id_,
    '2021' as year_,
    (select Name from headhunter.Account where employer_id == employer_id_) as employer_name_

select
    employer_id_,
    employer_name_,
    spending_year as report_year,
    sum(spending_cost)/sum(response_cnt) as sp_to_res_cost_avg,
    count(spending_id) as spending_cnt,
    sum(response_cnt) as response_cnt_,
    sum(spending_cost) as total_cost
from
(
select
    spending_id,
    toYear(spending_date) as spending_year,
    employer_service_id,
    vacancy_id,
    price_to_one as spending_cost,
    response_cnt,
    CASE
    WHEN response_cnt > 0 THEN price_to_one / response_cnt
    ELSE 0
    END as response_cost
from
    (
    select
        spending_id,
        spending_date,
        CASE
        WHEN spending_end_ == '1970-01-01 00:00:00' THEN now()
        ELSE spending_end_
        END as spending_end,
        employer_service_id,
        vacancy_id,
        price_to_one
    from(
        select
        id as spending_id,
        spending_date,
        leadInFrame(spending_date) OVER (ORDER BY spending_date ASC ROWS
                     BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS spending_end_,
        employer_service_id,
        vacancy_id
        from
            headhunter.Spending
        where
            vacancy_id in (select vacancy_id from headhunter.Vacancy where employer_id == employer_id_)
            and
            toYear(spending_date) == year_
        ) as l2
    left join
        (
        select
            adjusted_cost / original_cnt as price_to_one,
            employer_service_id
        from
            headhunter.Orders
        where
            employer_id == employer_id_
        ) as r2
    on l2.employer_service_id = r2.employer_service_id
    ) as l3
left join
    (
    select
        spending_id,
        count(ID) as response_cnt
    from
    (
    select
        *
    from
        (
        select
            spending_id,
            spending_date,
            CASE
            WHEN spending_end_ == '1970-01-01 00:00:00' THEN now()
            ELSE spending_end_
            END as spending_end,
            employer_service_id,
            vacancy_id,
            price_to_one
        from(
            select
            id as spending_id,
            spending_date,
            leadInFrame(spending_date) OVER (ORDER BY spending_date ASC ROWS
                         BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS spending_end_,
            employer_service_id,
            vacancy_id
            from
                headhunter.Spending
            where
                vacancy_id in (select vacancy_id from headhunter.Vacancy where employer_id == employer_id_)
                and
                toYear(spending_date) == year_
            ) as l
        left join
            (
            select
                adjusted_cost / original_cnt as price_to_one,
                employer_service_id
            from
                headhunter.Orders
            where
                employer_id == employer_id_
            ) as r
        on l.employer_service_id = r.employer_service_id
        ) as l1
    left join
        (
        select *
        from headhunter.VacancyResponse
        where
            vacancy_id in (select vacancy_id from headhunter.Vacancy where employer_id == employer_id_)
        ) as r1
    on l1.vacancy_id = r1.vacancy_id
    having
        creation_date >= spending_date
        and
        creation_date < spending_end
    )
    group by
        spending_id
    ) as r3
on l3.spending_id = r3.spending_id
)
group by
    spending_year

```
